Desafio 01 - Análise da Distribuição de Carros por Fabricante e Combustível


```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("CarsData.csv")

grouped_data = data.groupby(['Manufacturer', 'fuelType']).size().unstack()

grouped_data.plot(kind='bar', figsize=(12,6))
plt.title('Distribuição de Carros por Fabricante e Combustível')
plt.xlabel('Fabricante')
plt.ylabel('Número de Carros')
plt.legend(title='Combustível')
plt.show()

max_fuelType = grouped_data.idxmax(axis=1)
max_cars = grouped_data.max(axis=1)

max_fuelType_df = grouped_data.idxmax(axis=1).to_frame('combustível mais usado')

most_used_fuel = max_fuelType_df['combustível mais usado'].max()

min_fuelType_df = grouped_data.idxmin(axis=1).to_frame('combustível menos usado')

least_used_fuel = min_fuelType_df['combustível menos usado'].min()


print("\n**Análises levantadas***")

print(f"\n\nO combustível mais usado é o {most_used_fuel}")
print(f"O combustível menos usado é o {least_used_fuel}")

print("\n\nCombustíveis mais usados nos carros de cada fabricante:\n")
for manufacturer, fueltype in max_fuelType.items():
  print(f"\n{manufacturer}: {fueltype} (Número de carros que usa o tipo de combustível: {max_cars[manufacturer]})")
```
